```python
import csv
```

```python
roeroe = []
streets = {}
```

```python
with open('data-397-2019-02-02.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row['RepairOfEscalators'] != '':
            roeroe.append(row['Name'])
for stan in roeroe:
    if stan in streets:
        streets[stan] += 1
    else:
        streets[stan] = 1
```
